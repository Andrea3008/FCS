```python
def gb_mean(row):
    media = (row['ratings_1']*1 + row['ratings_2']*2 + row['ratings_3']*3 + row['ratings_4']*4 
             + row['ratings_5']*5)/row['work_ratings_count']
    return media
```

### 7. For each book in `goodbooks`, compute the list of its authors. Assuming that the number of reviews with a text (column `work_text_reviews_count`) is split equally among all authors, find for each authors the total number of reviews with a text. We will call this quantity the *shared number of reviews with a text*.

```python
def find_author(lista):
    final = list()
    for authors in lista:
        single_book_list = authors.split(', ')
        for author in single_book_list:
            final.append(author)
    return final
```

```python
def count_book_shared_number(row):
    shared_number = 0
    authors = row['authors'].split(', ')
    shared_number = row['work_text_reviews_count']/len(authors)
    return shared_number
```

```python
gb_shared = goodbooks[['isbn', 'authors', 'original_title', 'work_text_reviews_count', 'original_publication_year']]
gb_shared['shared_number'] = gb_shared.apply(count_book_shared_number, axis=1)
gb_shared.head()
```

```python
list_authors = find_author(goodbooks['authors'])
authors = pd.DataFrame(list_authors).drop_duplicates()
authors.columns = ['Name']
```

```python
def count_author_shared_number(row):
    books_by_author = gb_shared[gb_shared['authors'].str.contains(row['Name'])]
    list_shared_number = list(books_by_author['shared_number'])
    shared_number = 0
    for i in range(0, len(list_shared_number)):
        shared_number += list_shared_number[i]
    return shared_number
```

```python
complete_shared_number = authors.apply(count_author_shared_number, axis=1)
authors['Shared Number'] = complete_shared_number
authors.head()
```

### 10. Split the users dataset according to the age. One dataset contains the users with unknown age, one with age 0-14, one with age 15-24, one with age 25-34, and so on.

```python
users['Age'].max()
```

Il valore più alto di età è pari a 244, ciò vuol dire che vogliamo dividere il dataset users in 25 subset: il primo contenente le righe di utenti con età sconosciuta, il secondo con le righe di utenti con età in 0-14 e poi proseguire con gruppi di età di 9 anni in 9 anni (15-24, 25-34 ... 235-244).
Siccome le uniche categorie 'anomale' sono quelle in cui l'età è sconosciuta o è nel range 0-14, l'estrazione di queste viene considerata come 'caso base' mentre l'estrazione delle righe appartenenti alle categorie 'ordinarie', ovvero quelle che vanno di 9 anni in 9 anni, viene fatta automaticamente e i risultati vengono memorizzati in un dizionario. Per accedere ai singoli dataset contenuti nel dizionario basta indicare l'etichetta corrispondente all'intervallo inferiore del range di età:

ESEMPIO: si vuole accedere al dataset in cui ci sono le righe con età in 155-164 -> age_dict['155']

```python
age_null = pd.DataFrame()
age_0_14 = pd.DataFrame()
age_dict = {}
for x in range(15, int(users['Age'].max()), 10):
        age_dict[str(x)] = pd.DataFrame()
```

Siccome in users ci sono circa 279mila righe, i cicli for innestati richiedono un po' di tempo per produrre il risultato

```python
for index, row in users.iterrows():
    if pd.isnull(row['Age']):
        age_null = age_null.append(row)
    elif row['Age'] >= 0 and row['Age'] <= 14:
        age_0_14 = age_0_14.append(row)
    elif row['Age'] > 14:
        for j in range(15, int(users['Age'].max()), 10):
            if row['Age'] >= j and row['Age'] <= j+9:
                age_dict[str(j)] = age_dict[str(j)].append(row)
                break
```
